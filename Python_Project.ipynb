3. Bivariate Analysis: Exploring relationships between pairs of features.

```python
# Set plot style
sns.set(style="whitegrid")

# Define Categorical and Numerical Features
categorical_features = ['Warehouse_block', 'Mode_of_Shipment', 'Product_importance', 'Gender']
numerical_features = ['Customer_care_calls', 'Customer_rating', 'Cost_of_the_Product',
                      'Prior_purchases', 'Discount_offered', 'Weight_in_gms']

# 1. Categorical vs Target Variable
for feature in categorical_features:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=feature, hue='Reached.on.Time_Y.N', data=data, palette='coolwarm')
    plt.title(f"{feature} vs Reached.on.Time_Y.N")
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.legend(title="Reached.on.Time_Y.N", labels=["Not On Time", "On Time"])
    plt.xticks(rotation=45)
    plt.show()

# 2. Numerical vs Target Variable (Box Plots)
for feature in numerical_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Reached.on.Time_Y.N', y=feature, data=data, palette='coolwarm')
    plt.title(f"{feature} vs Reached.on.Time_Y.N")
    plt.xlabel("Reached on Time (1 = On Time, 0 = Not On Time)")
    plt.ylabel(feature)
    plt.show()
```

```python
# Encode categorical variables

encoded_data = data.copy()
encoded_data=encoded_data.drop(columns=['ID'])
for col in ['Warehouse_block', 'Mode_of_Shipment', 'Product_importance', 'Gender']:
    encoded_data[col] = LabelEncoder().fit_transform(encoded_data[col])

# Plot correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(encoded_data.corr(), annot=True, cmap="coolwarm", fmt=".2f", annot_kws={"size": 8})
plt.title("Correlation Heatmap")
plt.show()

# Plot correlation with target variable
correlations = encoded_data.corr()['Reached.on.Time_Y.N'].sort_values(ascending=False)
plt.figure(figsize=(8, 6))
correlations.drop('Reached.on.Time_Y.N').plot(kind='bar', color='steelblue')
plt.title("Correlation with Target (Reached.on.Time_Y.N)")
plt.show()
```

*ADVANCED INSIGHTS USING PREDICTIVE MODELLING*


```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

# Predictive Modeling: Logistic Regression
X = data[['Customer_care_calls', 'Customer_rating', 'Cost_of_the_Product', 
          'Prior_purchases', 'Discount_offered', 'Weight_in_gms']]
y = data['Reached.on.Time_Y.N']

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Logistic Regression Model
log_model = LogisticRegression(max_iter=1000, random_state=42)
log_model.fit(X_train, y_train)

# Predictions and Evaluation
y_pred = log_model.predict(X_test)
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Not On Time', 'On Time'], 
            yticklabels=['Not On Time', 'On Time'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Feature Importance
coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': log_model.coef_[0]
}).sort_values(by='Coefficient', ascending=False)
print("Feature Coefficients (Logistic Regression):")
print(coefficients)

# User Input Prediction
print("\nEnter the following values to predict if the order will reach on time:")
customer_care_calls = int(input("Number of Customer Care Calls: "))
customer_rating = int(input("Customer Rating (1-5): "))
cost_of_product = int(input("Cost of the Product: "))
prior_purchases = int(input("Number of Prior Purchases: "))
discount_offered = int(input("Discount Offered: "))
weight_in_gms = int(input("Weight of the Product (in grams): "))

# Create input array and make prediction
user_input = np.array([[customer_care_calls, customer_rating, cost_of_product, 
                        prior_purchases, discount_offered, weight_in_gms]])
prediction = log_model.predict(user_input)
probability = log_model.predict_proba(user_input)

if prediction[0] == 1:
    print(f"\nThe prediction is: The order WILL reach on time (Probability: {probability[0][1]:.2f})")
else:
    print(f"\nThe prediction is: The order WILL NOT reach on time (Probability: {probability[0][0]:.2f})")
```
